```python
import pandas as pd
import re
```

```python
df = pd.read_csv('C:\Python Data Science\MELBOURNE.csv')
```

```python
df
```

```python
filter1 = df.loc[df['Rooms'] ==3] # filtrando dados  especifico casa com 3 quartos
filter1
```

```python
filter2 = df.loc[df['Address'].str.contains('Turner St|Turner Rd', flags=re.I)] # filtrando nome de rua em especifico
filter2                                                                         # usando expressão regular, para filtrar letras maiasculas e minisculas
```

```python
filter3 = df.loc[df['Address'].str.contains('^59', flags=re.I) | (df['Price'] <= 500000)] # filtrando casas que começam com o numero 59
                                                                                          #ou casas abaixo de 500000
filter3
```
