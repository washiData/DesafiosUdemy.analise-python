```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('C:\gas.csv')
gas
```

```python
plt.plot(gas['Year'], gas['Australia']) # plotando coluna em especifico
```

```python
plt.title('Valor do Combustivel em US$')# adicionando titulo

plt.plot(gas['Year'], gas['Australia'], label = 'Australia')# label adicionado legenda ao grafico em especfico
plt.plot(gas['Year'], gas['Italy'], label = 'Italy')
plt.plot(gas['Year'],gas['USA'], label = 'USA')
plt.plot(gas['Year'],gas['Japan'], label = 'Japan')

plt.xticks(gas['Year'] [:: 2])#[:: 2]) isso siginifica do inicio ao fim em dois em dois anos 

plt.legend()
plt.show()
```
